### Sample data analysis

**Question:** Can we predict the number of views or answers to a newly posted question on Stack Overflow based on:
- the reputation of the owner,
- popularity of post's tags.

**Reasoning:** It is reasonable to assume that a more distinguished user and/or a question on a "hot" topic may generate more interest. We want to verify this assumption, to test if we can use this information, which is available at the time when a question is posted, to predict the interest in the question.

**Data source:** Stack Overflow has an SQL based data explorer [online](https://data.stackexchange.com/stackoverflow/queries).

- We only retrive posts from a certain time period.

```
declare @startTime date = '2015-08-01'
declare @endTime date ='2015-10-01'
```

 - Tags of a post: for a given post, we compute the number of tags, the frequency of the most popular post's tag, and the total number of posts which have at least on tag in common with the given post (incurs overcounting).
 
 ```
 create table #Total_Tags_Count (postId int, numTags int, maxTagCount int, totalTagCount int)
insert #Total_Tags_Count
select
     Posts.Id,
     Count(*),
     Max(Tags.Count),
     Sum(Tags.Count)
 from Posts
     inner join PostTags on PostTags.PostId = Posts.Id
     inner join Tags on Tags.Id = PostTags.TagId   
 where  Posts.ParentId is null
 group by Posts.Id
```

- Post information: finally, we combine the information about tags of a post with other information about the post, such as score, number of views, number of answers and comments, and reputation of the user who posted the question

```
select
   Posts.Id,
   Posts.Score as postScore,
   Posts.ViewCount as views,
   Posts.AnswerCount as answers,
   Posts.CommentCount as comments,
   totalTagCount,
   maxTagCount,
   numTags,
   Users.Reputation as ownerRep
 from Posts
   inner join Users on Users.Id = Posts.OwnerUserId
   inner join  #Total_Tags_Count as ttc on ttc.postId = Posts.Id
 where  Posts.ParentId is null
 and Posts.CreationDate>=@startTime
 and Posts.CreationDate<@endTime
```

Scatter plot did not provide much information. Next, we want to make several heat plots. The following function takes in as input variables:

- a database,
- two explanatory variables and one response variable,
- number of bins for each of explanatory variables,
- a function which is to be applied on data.

The function splits the explanatory variables into mutually disjoint, exhaustive, equal length intervals, whose number is given on input, to produce two dimensional bins. Then it groups the responses into these bins and applies a given function on data in each bin. It records the output in a matrix. 

````python
# prepare matrix for heat plot
def get_matrix(db, x,y,response, func, x_numbins, y_numbins):
    mat = [[0 for i in range(x_numbins)] for j in range(y_numbins)] #output matrix
    
    # determine interval length and bounds
    xstep = float(max(db[x])-min(db[x]))/x_numbins
    xbounds=[min(db[x])+i*xstep for i in range(x_numbins+1)]
    ystep = float(max(db[y])-min(db[y]))/y_numbins
    ybounds=[min(db[y])+j*ystep for j in range(y_numbins+1)]
    
    # group response variable in bins 
    for i in range(x_numbins):
        for j in range(y_numbins):
            ans = db[response][ (db[x]>=xbounds[i]) & (db[x]<xbounds[i+1]) & (db[y]>=ybounds[j]) & (db[y]<ybounds[j+1])]
            if ans.shape[0]>0: # if there are any answers in this category, apply the function to it
                mat[y_numbins-1-j][i] = func(ans)
    return [mat,xbounds,ybounds]     
````

Now, we will use this function to plot mean and median of the logarithm of views for posts which have total frequency of their tags and the log of owner's reputation within given bounds. We also count the number of posts in these categories. 

````python
xbins = 5
ybins = 4
mat1,xbounds,ybounds = get_matrix(db2, 'log_ownerRep','totalTagCount',  'log_views', np.mean, xbins,ybins)
ax1 = sns.heatmap(mat1, annot=True, fmt='f')

plt.title('Mean of log of views')
plt.xlabel('Log of owner\' reputation')
plt.ylabel('Total frequency of post\'s tags')

xlabels = ['%.2f'%xbounds[i] for i in range(xbins+1)] # format x-axis labels
plt.xticks(range(xbins+1), xlabels, rotation=0)
ylabels = ['%.d'%ybounds[i] for i in range(ybins+1)] # y-axis labels
plt.yticks(range(ybins+1), ylabels, rotation=0)

plt.margins(0.2) 
#plt.subplots_adjust(bottom=0.15) # spacing at the bottom
````
